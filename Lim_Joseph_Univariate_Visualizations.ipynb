## Countplots

Displaying a countplot of the number of passengers that died vs. those that survived.

```python
#Creating list for countplot later on for interpretability
survival_labels = ["No", "Yes"]
```

### Seaborn Plot

```python
#Creating initial count plot in seaborn
ax = sns.countplot(data = df, x = "Survived")

#Changing x-tick labels so they reflect if a passenger survived or not
ax.set(xticks=range(len(survival_labels)), xticklabels=[i for i in survival_labels])

#Creating title
ax.set_title("Non-Survivors vs. Survivors")

ax
```

### Interpretation

From this graph, we can see that more passengers on the Titanic died than survived. We can also see that roughly 1.5 times more passengers died than survived.

The matplotlib library does not provide an aesthetically pleasing count plot, so no matplotlib plot is provided.


## Histograms

### Seaborn Plots

```python
#Creating initial histogram
ax = sns.histplot(data = df, x = "Age")
ax.set_title("Age distribution of the Titanic")
```

```python
#Creating modified histogram for every 10 years
ax = sns.histplot(data = df, x = "Age", bins = (80//10))

ax.set_title("Age distribution of the Titanic")
```

### Matplotlib

```python
#Base matplotlib histogram
ax = df["Age"].hist()

ax.set_xlabel("Age bins")
ax.set_ylabel("Counts")
ax.set_title("Age Distribution of Passengers")
```

```python
#Better matplotlib histogram
ax = df["Age"].hist(bins = 80//10, edgecolor = "black")

ax.set_xlabel("Age Bins")
ax.set_ylabel("Counts")
ax.set_title("Age Distribution of Passengers")
ax.grid(False)
```

### Interpretation

We can see from these graphs that there were more 0-10 year olds than any other age group (10-20, etc.). There were less 70-80 year olds than any other age group.

## Boxplots

### Seaborn Plots

```python
ax = sns.boxplot(data = df, y = "Fare")

ax.set_title("Distribution of Fare")
ax.set_ylabel("Fare($)")
```

### Matplotlib Plots

```python
fig, ax = plt.subplots()

ax.boxplot(df["Fare"])

ax.set_xlabel("Fare")
ax.set_ylabel("Fare($)")
ax.set_title("Distribution of Fare")

ax
```

### Interpretation

We can see from our box plots that most passengers did not pay more than \$100 for their fare. We can also see that a small amount of passengers paid more than ~$80, up to over $500 per ticket.
