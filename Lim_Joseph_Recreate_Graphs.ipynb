## Data Importation

```python
#DataFrame manipulation libraries
import numpy as np
import pandas as pd

#Visualization Packages
import matplotlib.pyplot as plt
```

```python
#Importing in data from link
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSdgsW4J8l9V8UlKLuwdYdqyemwkAuTHzS4s6plb-BQIJ8DhI8BkNDvO_Ei3M4lBJElVEaueyOg1Jkj/pub?output=csv"

df = pd.read_csv(url)
df.head()
```

```python
fig, ax = plt.subplots()

x = df["Month"]
y3 = df["3% Cumulative Interest"]
y5 = df["5% Cumulative Interest"]

ax.plot(x, y3, label = "30 year 3% APR")
ax.plot(x, y5, label = "30 year 5% APR")

ax.legend()
ax.set_xlabel("Month")
ax.set_ylabel("Dollars ($)")
ax.set_title("Cumulative Interest on 400K Loan")
```
