```python
# load data - change the path to match your drive!
df_details = pd.read_csv("School_Details.csv")
# make sure it loaded correctly
df_details.head(2)
```

## Problems:

```python
#What is the average BACDistance (in Km)? Rename the output column to "avg_bac_distance."
# Write your SQL query in this string
sql = """
SELECT AVG(`BACDistance (in KM)`) as avg_bac_distance
FROM school_details
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# What is the average BACDistance (in Km) by School Level? 
# Sort the results from highest average to lowest average.

sql = """
SELECT AVG(`BACDistance (in KM)`) as avg_bac_distance_by_level, SchoolLevel
FROM school_details
GROUP BY SchoolLevel
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Repeat the previous query, but only display results 
#where the average distance is at least 10 km.
sql = """
SELECT AVG(`BACDistance (in KM)`) as avg_bac_distance_by_level, SchoolLevel
FROM school_details
GROUP BY SchoolLevel
HAVING avg_bac_distance_by_level > 10
;
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
#Join the two tables together (school_details and school_facility_details). 
#Start by displaying all of the columns, but 
#limit the resulting rows to just 2.
sql = """
SELECT *
FROM school_details d INNER JOIN school_facility_details f 
ON d.EstablishmentCode = f.EstablishmentCode
LIMIT 2
;
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Select just the schools where the SchoolLevel is "PS" that do not have libraries.
sql = """
SELECT *
FROM school_details d INNER JOIN school_facility_details f 
ON d.EstablishmentCode = f.EstablishmentCode
WHERE SchoolLevel IS "PS" AND IsLibraryAvailable IS "Available"
;
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Order the results alphabetically by name.
sql = """
SELECT *
FROM school_details d INNER JOIN school_facility_details f 
ON d.EstablishmentCode = f.EstablishmentCode
WHERE SchoolLevel IS "PS" AND IsLibraryAvailable IS "Available"
ORDER BY EstablishmentName ASC
;
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```
