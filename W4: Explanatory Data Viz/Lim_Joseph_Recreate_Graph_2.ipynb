```python
import pandas as pd
import seaborn as sns
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vS8EOvOoG2CKLd7KBQ2fetTPXDBioP3qjzzf45xjxxyDlC_WLeZdXi_k8r9UkWsRd9KiyPQXTwchi1C/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python
min_bmi = 18.5
max_bmi = 24.9
ax = sns.histplot(data = df, x = "bmi")
ax.set_title("bmi")

ax.axvline(min_bmi, color = "red", label = f"Min Recommended BMI: {min_bmi}")
ax.axvline(max_bmi, color = "red", label = f"Max Recommended BMI: {max_bmi}")
ax.legend()
```
