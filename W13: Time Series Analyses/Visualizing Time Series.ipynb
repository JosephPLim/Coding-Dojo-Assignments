# Visualizing Time Series

## Imports and Libraries

```python
#Dataframe manipulation
import pandas as pd

#Visualization stuff
import matplotlib.pyplot as plt

#import tick customization tools
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
```

```python
#Setting default parameters, 12 long by 3 high
plt.rcParams["figure.figsize"] = (12,3)
```

```python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vT_jChgNsQbHbg4TGepzIqk8XC9DTIKmyyxb1upo5cfZCgbfIUQc2ZC0YMzuU5uApP140Ob49KBjdqh/pub?gid=1198589591&single=true&output=csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
# Converting data column to datetime datatype
df["date"] = pd.to_datetime(df["date"], format = "%Y%m%d")

#Checking if data is correct
df.head()
```

```python
#Checking if type of feature is correct
df["date"].info()
```

```python
#Setting date as index
df = df.set_index("date")
df.head()
```

Frequency of data appears to be daily, including holidays

```python
# Keeping data from 2000 and later
df = df.loc["2000":]
df.head()
```

```python
#Creating a set amount of features
feature_list = ["precipitation", "mean_temp", "min_temp",
               "max_temp", "snow_depth"]
```

```python
#Only keeping the set features
df = df[feature_list]
df.head()
```

```python
#Checking for null values
df.isna().sum()
```

```python
# Filling "precipitation" with 0 values, assuming no rain on those days
df["precipitation"] = df["precipitation"].fillna(value = 0)
df["precipitation"].isna().sum()
```

```python
# All other values can be interpolated
df = df.interpolate()
df.head()
```

```python
df.isna().sum()
```

## Data Visualization

```python
data_viz_df = df.copy()
data_viz_df = data_viz_df.loc["2018":"2020"]
data_viz_df = data_viz_df[["mean_temp", "min_temp", "max_temp"]]
data_viz_df
```

```python
title = "Daily Temperature"
ylabel = "Degrees"
```

```python
## creating our tick locators and formatters
## for minor month ticks
# create the locator to place ticks every 3 months.
loc_3months = mdates.MonthLocator(interval=3)
fmt_months = mdates.DateFormatter("%b")
## for major year ticks
loc_year = mdates.YearLocator()
fmt_year = mdates.DateFormatter("%Y")
```

```python
# Plotting 2018-2020, for min, mean, and max temperatures
fig, ax = plt.subplots(figsize=(12,3))
data_viz_df.plot(ax=ax,title=title);
##  Set xticks and xticklables
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), fontsize='large',fontweight='bold' ) 
##  customize minor ticks
ax.xaxis.set_minor_locator(loc_3months)
ax.xaxis.set_minor_formatter(fmt_months)
##  customize major ticks
ax.xaxis.set_major_locator(loc_year)
ax.xaxis.set_major_formatter(fmt_year)
##  Making major/minor gridlines visually distince
ax.grid(which='minor',axis='x',ls=":")
ax.grid(which='major',axis='x',color='k')

ax.set_ylabel(ylabel)
fig.autofmt_xdate(rotation=90,ha='center')
```
