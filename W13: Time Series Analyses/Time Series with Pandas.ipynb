## Working with Data

```python
df.info()
```

```python
df.head()
```

```python
# Converting column "date" to a datetime object
df["date"] = pd.to_datetime(df["date"], format = "%Y%m%d")
display(df.head())
df.info()
```

```python
# Setting date column as index
df = df.set_index("date")
df.head()
```

```python
# Keeping data from 2000 onwards
df = df.loc["2000":]
df.head()
```

```python
#Creating a set amount of features
feature_list = ["precipitation", "mean_temp", "min_temp",
               "max_temp", "snow_depth"]
```

```python
#Only keeping the set features
df = df[feature_list]
df.head()
```

### When was the lowest temperature within this dataset?

```python
# Getting lowest temperature within dataset
min_temp = df["min_temp"].idxmin()

# Getting data associated with lowest temperature within dataset
precip, mean, minim, maxim, snow = df.loc[min_temp]

df.loc[min_temp]
```

```python
# Getting time data related to lowest temperature within dataset
date = min_temp.strftime("%B %d, %Y")
day = min_temp.strftime("%A")

#Printing Answer
print(f"The lowest temperature of {minim} degrees occurred on {date} which was a {day}.")
```

### When was the highest precipitation within this dataset?

```python
# Getting highest precipitation within dataset
max_precip = df["precipitation"].idxmax()

# Getting data associated with lowest temperature within dataset
precip, mean, minim, maxim, snow = df.loc[max_precip]

df.loc[max_precip]
```

```python
# Getting time data related to maximum precipitation within dataset
date = max_precip.strftime("%B %d, %Y")
day = max_precip.strftime("%A")

#Printing Answer
print(f"The highest daily precipitation of {precip} inches occurred on {date} which was a {day}.")
```
